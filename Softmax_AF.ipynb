```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sfm(x):
    return np.exp(x-np.max(x))/np.sum(np.exp(x-np.max(x)))
```

```python
x=np.array([2.0,1.0,0.1])
y=sfm(x)
```

```python
plt.bar(range(len(y)),y,tick_label=['c1','c2','c3'])
plt.ylabel('probability')
plt.ylim(0,1)
print("softmax probs:", y)
print("probs sum:", np.sum(y))
```
