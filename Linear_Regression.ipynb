```python
from sklearn.linear_model import LinearRegression # LinearRegression class to perform Linear Regression
import numpy as np # used to work with neumeric data
import matplotlib.pyplot as plt # to visualize the data and regression line
```

```python
x=np.array([[1000],[1500],[2000],[2500],[3000]]) # input data as 5 row and 1 columns
# 2D array has been used because sklearn expects inputs as rows and columns
# mainly to store multiple features as matrix

y=np.array([200,250,300,350,400]) # target variables
# here 1D array is used as 1 output is enough for each input 
```

```python
# creating a linear regression model
model=LinearRegression()
```

```python
# training the model using data
model.fit(x,y) # fit the model: finds the best lines to map x and y 
```

```python
# predicting the house price of size 1800 sq ft
pdt_price=model.predict([[1800]])
print("predicted price: ",pdt_price)
```

```python
# visualizing the data and the regression line
plt.scatter(x,y, color='blue') # prints the blue dots
plt.plot(x,model.predict(x),color='red') # connects with red line
plt.grid() # displays the grid
```
