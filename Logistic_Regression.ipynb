```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# input data
x=np.array([[1],[2],[3],[4],[5],[6]]) # 2D array containing 1 feature: hours studied

# target values
y=np.array([0,0,0,1,1,1]) # 1D array for binary classification
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
pdt=model.predict([[3.5]]) # predicts class lable 0 or 1
prob=model.predict_proba([[3.5]]) # predicts the probability of both the classes
```

```python
print("prediction", pdt[0])
print("probability ", prob[0]) # [fail,pass]
```

```python
plt.scatter(x, y, color='blue', label='training data') # plots the blue points
plt.plot(x,y,color='red',label='logistic regression') # connects the dotes with red line
plt.legend() # show legend
plt.grid() 
```
