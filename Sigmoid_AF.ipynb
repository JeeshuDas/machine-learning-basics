```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def smd(x):
    return 1/(1+np.exp(-x)) # formula for sigmoid
```

```python
x=np.linspace(-10,10,100) # 100 points between -10 and +10
y= smd(x)
```

```python
plt.plot(x,y,label='sigmoid')
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
```
