```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([
    [1, 2],
    [1.5, 1.8],
    [5, 8],
    [8, 8],
    [1, 0.6],
    [9, 11],
    [8, 2],
    [10, 2],
    [9, 3]
])
model=KMeans(n_clusters=3)
model.fit(x)
```

```python
cn=model.cluster_centers_
# this line gives the coordinates of the centroid

lb=model.labels_
# gets the labels for each datapoint

cl=['r','g','b']
```

```python
for i in range(len(x)):
    plt.scatter(x[i][0], x[i][1], color=cl[lb[i]])
plt.scatter(cn[:,0],cn[:,1],marker="o",s=200,linewidths=3,color='black',label='centroids')
plt.legend()
plt.grid(True)
```
