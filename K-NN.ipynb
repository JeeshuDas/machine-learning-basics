```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[1,2],[2,3],[3,1], # class 0 points
            [6,5],[7,7],[8,6]]) # class 1 points
# the x and y coordinates as [x,y] will show a point in graph
# class 0 is the closest points to origin and class 1 is the farthest points from the origin

y=np.array([0,0,0,1,1,1]) # target variables
# showing first 3 inputs belong to class 0 and next 3 inputs belong to class 1 
```

```python
model=KNeighborsClassifier(n_neighbors=3) # n=3 makes the model to look for 3 nearest datapoints for analysis
```

```python
model.fit(x,y)
```

```python
p1=np.array([[4.4,4.4]])
pdt=model.predict(p1)
print(pdt)
```

```python
plt.scatter(x[:3,0],x[:3,1],color='blue') # class 0 points in blue
plt.scatter(x[3:,0],x[3:,1],color='red') # class 1 points in red
plt.scatter(p1[0,0],p1[0,1],color='black', marker='X',s=100)
plt.grid()
```
