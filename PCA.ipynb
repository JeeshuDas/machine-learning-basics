```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
x=np.array([
    [160, 55],
    [165, 58],
    [170, 63],
    [175, 67],
    [180, 70],
    [185, 75]
])
sc=StandardScaler()
x_sc=sc.fit_transform(x)
```

```python
# reducing 2D data to 1D data
pca=PCA(n_components=1)
x_pca=pca.fit_transform(x_sc)
```

```python
print(x_sc.shape)
print(x_pca.shape)
print(x_pca)
```

```python
plt.figure(figsize=(8,3))
plt.scatter(x_sc[:,0],x_sc[:,1],label="2d data",color='blue') # plot 2D original data points
```

```python
for i in range(len(x_sc)):
    plt.plot([x_sc[i, 0], pca.inverse_transform(x_pca[i])[0]],
             [x_sc[i, 1], pca.inverse_transform(x_pca[i])[1]],
             'r--')
plt.xlabel("F1")
plt.ylabel("F2")
```
