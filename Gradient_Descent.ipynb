```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
w=0
b=0
lr=0.01
epochs=1000
n=len(x)
```

```python
for i in range(epochs):
    pdt=w*x+b # predicted output
    er=y-pdt # error
    dw=(-2/n)*sum(x*er) # derivative wrt weight
    db=(-2/n)*sum(er) # derivative wrt bias
    w=w-lr*dw # new weight
    b=b-lr*db # new bias
    if i%100==0:
        loss=np.mean(er**2) # mean square error as loss
        print(f"Epoch {i}, loss: {loss:.4f}, w: {w:.4f}, b: {b:.4f}")
```

```python
print(f"trained model: y= {w:.2f}x+{b:.2f}")
```

```python
plt.scatter(x,y,color='blue',label='actual') # original datapoints
plt.plot(x,w*x+b,color='red',label='fitted line') # predicted line
plt.legend()
plt.grid()
```
